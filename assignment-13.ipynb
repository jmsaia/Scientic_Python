```python
# Use the Matplotlib object-oriented interface to display the sine, cosine,
# tangent, and arctangent of x, each in its own plot.

x = np.linspace(0, 10, 100)

fig, ax = plt.subplots(4, figsize=(12,12))

ax[0].plot(x, np.sin(x))
ax[1].plot(x, np.cos(x))
ax[2].plot(x, np.tan(x))
ax[3].plot(x, np.arctan(x));
```

## Requirement 17

```python
# Produce the following plot (purple and orange). This page will be helpful:
  
plt.style.use('seaborn-white')

def f(x, y):
    return np.sin(x) ** 10 + np.cos(10 + y * x) * np.cos(x)

x4 = np.linspace(0, 5, 50)
y4 = np.linspace(0, 5, 40)

X, Y = np.meshgrid(x4, y4)
Z = f(X, Y)

contours = plt.contour(X, Y, Z, 3, colors='black')
plt.clabel(contours, inline=True, fontsize=8, fmt='%.3f')

plt.imshow(Z, extent=[0, 5, 0, 5], origin='lower',
           cmap='PuOr', alpha=0.5)

plt.colorbar();
```

## Requirement 19

```python
# Produce the following plot (notice the color and legends)

plt.style.use('classic')

fig, ax = plt.subplots()

lines = []
styles = ['-', '--', '-.', ':']
x = np.linspace(0, 10, 1000)

for i in range(4):
    lines += ax.plot(x, np.sin(x - i * np.pi / 2),
                     styles[i], color='green')
ax.axis('equal')

# specify the lines and labels of the first legend
ax.legend(lines[:2], ['line A', 'line B'],
          loc='upper left', frameon=True)

# Create the second legend and add the artist manually.
from matplotlib.legend import Legend
leg = Legend(ax, lines[2:], ['line C', 'line D'],
             loc='lower left', frameon=True)

ax.add_artist(leg);
```
